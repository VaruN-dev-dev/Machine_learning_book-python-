## scatter plot

```python
flowers_df = sns.load_dataset("iris")
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
# from matplotlib
plt.scatter(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
# using seaborn
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
# just see the changes between the two in this one we also got labels..
```

### adding hues

```python
# you can also include plt function here cause seaborn is built on top of matplotlib
# plt.figure(figsize=(15, 10))
plt.title("sepal dimenstions")
sns.scatterplot(x=flowers_df.sepal_length,
               y=flowers_df.sepal_width,
               hue=flowers_df.species, s=100)
# the hue here is something like .. it is same length as that or the other one's
# it just gives different colours based on different species
# s here is for size
```

## plotting using pandas data frame

```python
plt.title("sepal dimenstions")
sns.scatterplot(x="sepal_length",
               y="sepal_width",
               hue="species",
               s=100,
               data=flowers_df)
```

# histogram

```python
plt.title("distribution of sepal width")
# plt.hist(flowers_df.sepal_width)
# plt.hist(flowers_df.sepal_width, bins=5)
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
# we can also make unequal bins...
```

## multiple histograms

```python
setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.legend(["setosa", "versicolor"])
```

```python
# we can also stack multiple histogram on top on one another
```

```python
plt.title("distribution of sepal width")

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, 
          virginica_df.sepal_width], 
          bins=np.arange(2, 5, 0.25),
         stacked=True);
plt.legend(["setosa", "versicolor", "virginica"])
```
